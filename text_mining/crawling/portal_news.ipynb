```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pymysql
import time
```

```python
def connectDB(dbData):
    DB_HOST = '127.0.0.1'
    DB_USER = 'root'
    DB_PASSWD = 'autoset'
    DB_NAME = 'python'
    conn = pymysql.connect(host=DB_HOST, user=DB_USER, password=DB_PASSWD,
                       db=DB_NAME, charset='utf8')
    curs = conn.cursor()
    sql = """insert into portalnews(url, title, date, sub_title, contents)
         values (%s, %s, %s, %s, %s)"""
    curs.executemany(sql,dbData)
    conn.commit()
    conn.close()
```

```python
def portal(search):
    url = "http://www.portalnews.co.kr/news/search_result.html?search={search}".format(search=search)
    driver = webdriver.Chrome("./chromedriver.exe")
    driver.get(url)
    return portalUrlList(driver, search)
```

```python
def portalInfo(driver, portalUrls_List):
    
    for urlList in portalUrls_List:
        driver.get(urlList)
        time.sleep(2)
        try:
            
            newsTitle = ""
            newsTitleList = driver.find_elements_by_css_selector(".art_top")
            newsList = [news.text.split("\n") for news in newsTitleList]
            newsList = [news for newslist in newsList for news in newslist]
            
            try:
                if len(newsList) == 11:
                    newsTitle_temp = newsTitleList[0]
                    newsTitle = newsTitle_temp.text.split("\n")[0]

                    newsSubtitle = ""
                    newsSubtitleList = driver.find_elements_by_css_selector(".art_top")
                    newsSubtitle_temp = newsSubtitleList[0]
                    newsSubtitle = newsSubtitle_temp.text.split("\n")[1]

                    newsDate = ""
                    newsDateList = driver.find_elements_by_css_selector(".art_top")
                    newsDateList_temp = newsDateList[0]
                    newsDate = newsDateList_temp.text.split("\n")[3]
                    newsDate = newsDate.replace("등록", "").replace(".", "")[0:9]

                    newsContents = ""
                    newsContentsList = driver.find_elements_by_css_selector("#news_body_area.smartOutput")
                    for newsContent in newsContentsList:
                        newsContents = newsContents + newsContent.text
                        newsContents = newsContents.replace("(포탈뉴스)", "")
                    print("Url : ",  urlList)
                    print("제목 : " , newsTitle)
                    print("부제목 : ", newsSubtitle)
                    print("날짜 : ", newsDate)
                    print("-- 내용 --")
                    print(newsContents)
                    
                    dbdata = [[urlList, newsTitle, newsDate, newsSubtitle, newsContents]]
                    connectDB(dbdata)
            except:
                continue
        except:
            continue
        
        
```

```python
def portalUrlList(driver, search):
    
    for pageNo in range(1, 11):
        url = "http://www.portalnews.co.kr/news/search_result.html?search=%EC%BD%94%EB%A1%9C%EB%82%98&page={pageNo}".format(pageNo = pageNo, search = search)
        driver.get(url)
        time.sleep(2)
        
        portalUrls_List = []
        portalUrl = driver.find_elements_by_css_selector(".art_list_all a")
        for url in portalUrl:
            portalUrls_List.append(url.get_attribute("href"))
        
        portalInfo(driver, portalUrls_List)
```

```python
search = "코로나19"
portal(search)
```
