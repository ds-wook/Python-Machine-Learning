```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
plt.style.use('ggplot')
```

```python
cat = pd.read_csv("./data/CatStory_survey.csv")
cat.head()
```

# 사용자 정보

```python
fig = plt.figure(figsize = (10, 10))
x = cat['1. 사용자 연령은 어떻게 되시나요?'].value_counts()
labels = cat['1. 사용자 연령은 어떻게 되시나요?'].unique()
explode = [0.01 for i in range(len(labels))]
plt.pie(x = x, labels = labels, autopct='%0.1f%%', colors = ["lightcoral", "skyblue", "limegreen", "palegoldenrod"], explode = explode)
plt.title('1. 사용자 연령은 어떻게 되시나요?', fontsize=20)
plt.legend(loc = "best")
plt.show()
```

```python
fig = plt.figure(figsize = (10, 10))
x = cat['2. 사용자의 성별은 어떻게 되시나요?'].value_counts()
labels = cat['2. 사용자의 성별은 어떻게 되시나요?'].unique()
explode = [0.01 for i in range(len(labels))]
plt.pie(x = x, labels = labels, autopct='%0.1f%%', colors = ["lightcoral", "skyblue", "limegreen", "palegoldenrod"], explode = explode)
plt.title('2. 사용자의 성별은 어떻게 되시나요?', fontsize=20)
plt.legend(loc = "best")
plt.show()
```
