```python
string = '4-1. 4번에 예를 체크했다면 어떤 조작이 어려웠는지 골라주세요. (중복 가능)'
data = [i for s in cat[string].fillna("null") for i in s.split(", ") if s != "null"]
df = pd.DataFrame(columns = [string], data = data)
fig = plt.figure(figsize = (10, 10))
ncount = len(cat[~cat[string].isna()])
ax = sns.countplot(y = string, data = df, order = df[string].value_counts().index)
ax2 = ax.twinx()
ax2.xaxis.tick_top()
ax.xaxis.tick_bottom()

ax.xaxis.set_label_position("bottom")
ax2.xaxis.set_label_position("top")
ax2.set_ylabel("Frequency[%]")
for p in ax.patches:
    x = p.get_bbox().get_points()[1, 0]
    y = p.get_bbox().get_points()[:, 1]
    ax.annotate("{}({:.1f} %)".format(int(x), 100 * x / ncount), (x, y.mean()), ha = "left", va = "center")
plt.title(string, fontsize = 25)
plt.show()
```
